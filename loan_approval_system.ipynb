```python
x = df.iloc[:, np.r_[1:5, 9:11]].values  # Removed 13:15 as it is out of bounds
y = df.iloc[:, 12].values  # This is valid since 12 is within 0-12
y
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

# Convert all categorical columns to string before encoding
x_train = x_train.astype(str)
x_test = x_test.astype(str)

# Apply LabelEncoder to the first 5 columns
for i in range(0, 5):
    x_train[:, i] = le.fit_transform(x_train[:, i])  # Fit on train
    x_test[:, i] = le.transform(x_test[:, i])  # Transform test

# Correct encoding for column 5
le_5 = LabelEncoder()
x_train[:, 5] = le_5.fit_transform(x_train[:, 5])  # Fit on train
x_test[:, 5] = le_5.transform(x_test[:, 5])  # Transform test
```

```python
x_train = x_train.astype(str)
x_test = x_test.astype(str)
```

```python
from sklearn.preprocessing import LabelEncoder
import numpy as np

le_y = LabelEncoder()
y_train = le_y.fit_transform(y_train)  # Fit on training labels

# Ensure 'UNKNOWN' is part of the classes to avoid errors
known_classes = list(le_y.classes_) + ['UNKNOWN']
le_y.classes_ = np.array(known_classes)  # Manually set classes

# Replace unseen labels with 'UNKNOWN' before transforming
y_test = [label if label in le_y.classes_ else 'UNKNOWN' for label in y_test]
y_test = le_y.transform(y_test)  # Now transform safely
y_test
```

```python
y_train = y_train.astype(str)
y_test = y_test.astype(str)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier()
classifier.fit(x_train, y_train)
```

```python
from sklearn import metrics
y_pred = classifier.predict(x_test)
print(metrics.accuracy_score(y_test, y_pred))
y_pred
```
